```python
price_df['lb'] = price_df['center'] - 2 * price_df['Adj Close'].rolling(20).std()
price_df.iloc[18:25]
```

```python
n = 20
sigma = 2
```

```python
def bollinger_band(price_df, n, sigma):
    bb = price_df.copy()
    bb['center'] = price_df['Adj Close'].rolling(n).mean()
    bb['ub'] = bb['center'] + sigma * price_df['Adj Close'].rolling(n).std()
    bb['lb'] = bb['center'] - sigma * price_df['Adj Close'].rolling(n).std()
    return bb
```

```python
bollinger = bollinger_band(price_df, n, sigma)
```

```python
def tradings(sample, book):
    for i in sample.index:
        if sample.loc[i, 'LAST'] > sample.loc[i, 'ub']:
            book.loc[i, 'trade'] = ''
        elif sample.loc[i, 'lb'] > sample.loc[i, 'LAST']:
            if book.shift(1).loc[i, 'trade'] == 'buy':    #  
                book.loc[i, 'trade'] = 'buy'     #  
            else:
                book.loc[i, 'trade'] = 'ready'    # 
        elif sample.loc[i, 'ub'] >= sample.loc[i, 'LAST'] and sample.loc[i, 'LAST'] >= sample.loc[i, 'lb']:
            if book.shift(1).loc[i, 'trade'] == 'buy' or book.shift(1).loc[i, 'trade'] == 'ready':
                book.loc[i, 'trade'] = 'buy'
            else:
                book.loc[i, 'trade'] = ''
    return (book) 
```
